### metacharacters 3: quantifiers

* {n} matches exactly n times
*{n,m} mathches at least n times, but no more than m times
* {n,} matches at least n times, but maybe infinite times
* "+" match at least one ({1,})
* "*" match zero or more times
* ? match one time or zero times


```python
[line for line in subjects if re.search(r"[A-Z]", line)]
```

```python
[line for line in subjects if re.search(r"[A-Z]{10}", line)]
```

```python
[line for line in subjects if re.search(r"[A-Z]{15,}", line)]
```

```python
[line for line in subjects if re.search(r"[aeiou]{4}", line)]
```

```python
[line for line in subjects if re.search(r"^F[wW]d?:", line)]
```

```python
[line for line in subjects if re.search(r"[nN]ews.*!$", line)]
```

```python
[line for line in subjects if re.search(r"^R[eE]:.*\b[iI]nvestor", line)]
```

### more metacharacters: alternation

* (?:x|y) match eihter x or y
* (?:x|y|y) match x, y, or z

```python
[line for line in subjects if re.search(r"\b(?:[Ccat]|[kK]itty)", line)]
```

```python
[line for line in subjects if re.search(r"\b(?:[Cc]at|[kK]itty)", line)]
```
