```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Billionaire Characteristics Database/richpeople.xlsx")
```

```python
recent = df[df['year'] == 2014]
```

```python
recent['countrycode'].value_counts().head(5) #499 many bil
```

```python
recent.sort_values(by='networthusbillion', ascending=False).head(5)
```

```python
recent[recent['rank'] == 1565]
```

```python
recent['networthusbillion'].describe()
```

```python
recent[recent['gender'] == 'male'].describe()
```

```python
recent.groupby(recent['gender']).describe()
```

#Adding up wealth by country

```python
recent.groupby('countrycode')['networthusbillion'].sum().sort_values(ascending=False).head(10)
```

#Industries

```python
recent['industry'].value_counts()
```

#whats the total amount of billionaire money from each industry

```python
x = recent.groupby('industry')['networthusbillion'].sum().sort_values(ascending = False)
```

```python
x.plot(kind='barh', x='name', y='networthusbillion', legend = False)
```

```python
# YOung and rich and cool: Billionaires under 35:
sorted_young_bils = recent[recent['age'] < 35].sort_values(by='networthusbillion', ascending = True)
#recent[recent['age'] < 40].plot(kind='barh', x='name', y='networthusbillion')
sorted_young_bils.plot(kind='barh', x='name', y='networthusbillion', legend = False)
```
