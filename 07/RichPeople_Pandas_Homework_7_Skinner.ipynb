```python
female_billionaires = recent[recent['gender'] == 'female']
male_billionaires = recent[recent['gender'] == 'male']
print("Female billionaire average wealth:", female_billionaires['networthusbillion'].mean())
print("Male billionaire average wealth:", male_billionaires['networthusbillion'].mean())

females = female_billionaires['networthusbillion'].mean()
males = male_billionaires['networthusbillion'].mean()
difference = ((float(females) - float(males)) * 1000000000)
females_more_than_males = int(difference) / 1000000
```

```python
print("Female billionaires are slighthly richer than their mail counterparts. In average they have", int(females_more_than_males), "Million US-Dollars more than men." )
```

```python
total_billions_women = int(female_billionaires['networthusbillion'].sum())
```

```python
total_billions_men = int(male_billionaires['networthusbillion'].sum())
```

```python
print("In total, however, male billions have", int(total_billions_men / total_billions_women), "times more billions than women. In total:",total_billions_men, "billion US-Dollars. That is more than the GDP of France.")
```
