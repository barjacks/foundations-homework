## 4) How many of the richest Swiss are female, and how many are male?

So that is 8 female Swiss billionaires and 14 male ones.

## 5) How is their wealth distributed? How many billions do super rich men and women have in Switzerland compared to women?

```python
female_Swiss_billionaires_list = Swiss_billionaires[Swiss_billionaires['gender'] == 'female']
female_mean = female_Swiss_billionaires_list['networthusbillion'].mean()
print(female_mean)

#print("Swiss female billionaire's average wealth:", female_Swiss_billionaire['networthusbillion'].mean())
#print("Swiss male billionaire's average wealth:", male_Swiss_billionaire['networthusbillion'].mean())
```

```python
male_Swiss_billionaires_list = Swiss_billionaires[Swiss_billionaires['gender'] == 'male']
male_mean = male_Swiss_billionaires_list['networthusbillion'].mean()
print(male_mean)
```

```python
diff = male_mean - female_mean
Millions_diff = diff * 1000
male_sum = male_Swiss_billionaires_list['networthusbillion'].sum()
female_sum = female_Swiss_billionaires_list['networthusbillion'].sum()

print("In total Swiss male billionaires have funds of", male_sum, "billion US-Dollars. Female billionaires own", female_sum,"billion meaning that in average a Swiss male billionaire has", Millions_diff, "million USD more than Swiss female billionaires. That's qite a bit of cash!")
```

## 12. We compared women's and men's average fortunes before for Switzerland. Lets do the same for the whole world. 

```python
female_billionaires = recent[recent['gender'] == 'female']
male_billionaires = recent[recent['gender'] == 'male']
print("Female billionaire average wealth:", female_billionaires['networthusbillion'].mean())
print("Male billionaire average wealth:", male_billionaires['networthusbillion'].mean())

females = female_billionaires['networthusbillion'].mean()
males = male_billionaires['networthusbillion'].mean()
difference = ((float(females) - float(males)) * 1000000000)
females_more_than_males = int(difference) / 1000000
```

```python
print("Female billionaires are slighthly richer than their mail counterparts. In average they have", int(females_more_than_males), "Million US-Dollars more than men." )
```

```python
total_billions_women = int(female_billionaires['networthusbillion'].sum())
```

```python
total_billions_men = int(male_billionaires['networthusbillion'].sum())
```

```python
print("In total, however, male billions have", int(total_billions_men / total_billions_women), "times more billions than women. In total:",total_billions_men, "billion US-Dollars. That is more than the GDP of France.")
```
